```python
from openai import OpenAI

client=OpenAI(api_key="")
```

```python
training_file_path ='data/training_polymer_both_for_gpt.jsonl'
validation_file_path = 'data/validation_polymer_both_for_gpt.jsonl'
```

```python
training_file=client.files.create(file=open(training_file_path,'rb'),
                    purpose='fine-tune')

validation_file=client.files.create(file=open(validation_file_path,'rb'),
                    purpose='fine-tune')
print(training_file)
print('---------------------------------')
print(validation_file)
```

```python
response=client.fine_tuning.jobs.create(
    training_file=training_file.id,
    validation_file=validation_file.id,
    model='gpt-4o-mini-2024-07-18',
    hyperparameters={
    'n_epochs':'auto',
    'batch_size':'auto',  
    'learning_rate_multiplier': 0.1,
},
seed=42,
suffix="tsmp_monomers"
)

print(response)
```

```python
ft_model=client.fine_tuning.jobs.retrieve(response.id)#(response.id)
print(ft_model)
```
